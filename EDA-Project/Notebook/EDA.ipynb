Univariate Analysis
Numerical Columns

```python
numeric_cols = df.select_dtypes(include=np.number).columns

print(numeric_cols)
```

Histograms

```python
df[numeric_cols].hist(figsize=(12,8))
plt.tight_layout()
plt.show()
```

Boxplots

```python
for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

Categorical Analysis

```python
cat_cols = df.select_dtypes(include='object').columns

print(cat_cols)
```

```python
for col in cat_cols:
    print(df[col].value_counts())
    print("="*50)
```

```python
for col in cat_cols:
    plt.figure(figsize=(8,4))
    df[col].value_counts().head(10).plot(kind='bar')
    plt.title(col)
    plt.show()
```

Correlation Analysis

```python
corr = df.corr(numeric_only=True)

corr
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

Convert Order_Date

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month_name()

df.head()
```

Dataset Overview

```python
print("Rows and Columns:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())
```

Basic Statistics

```python
df[['Age','Quantity','Unit_Price','Total_Sales']].describe()
```

Sales Distribution

```python
plt.figure(figsize=(8,5))

df['Total_Sales'].hist(bins=20)

plt.title("Total Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

Gender Analysis

```python
df['Gender'].value_counts().plot(
    kind='bar'
)

plt.title("Customers by Gender")

plt.show()
```

City Analysis

```python
df['City'].value_counts().plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Customers by City")

plt.show()
```

Top 10 Products

```python
top_products = df.groupby(
    'Product'
)['Total_Sales'].sum().sort_values(
    ascending=False
).head(10)

top_products.plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Top 10 Products by Sales")

plt.show()
```

Category Revenue

```python
category_sales = df.groupby(
    'Category'
)['Total_Sales'].sum()

category_sales.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Revenue by Category")

plt.show()
```

Monthly Sales Trend

```python
monthly_sales = df.groupby(
    'Month'
)['Total_Sales'].sum()

monthly_sales.plot(
    marker='o',
    figsize=(10,5)
)

plt.title("Monthly Sales Trend")

plt.ylabel("Sales")

plt.show()
```

Top Cities by Revenue

```python
city_sales = df.groupby(
    'City'
)['Total_Sales'].sum().sort_values(
    ascending=False
)

city_sales.plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Revenue by City")

plt.show()
```

Correlation Analysis

```python
corr = df[
    ['Age',
     'Quantity',
     'Unit_Price',
     'Total_Sales']
].corr()

corr
```

Heatmap

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

Quantity vs Sales

```python
plt.figure(figsize=(7,5))

plt.scatter(
    df['Quantity'],
    df['Total_Sales']
)

plt.xlabel("Quantity")
plt.ylabel("Total Sales")

plt.title("Quantity vs Sales")

plt.show()
```
